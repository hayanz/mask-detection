```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
result = pd.read_csv("result_second.csv")
result.head()
```

```python
# loss of the train set at each epoch
# plt.plot(result['epoch'], result['train_cost'])
plt.plot(result['epoch'][1:], result['train_cost'][1:])
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("loss of the train set")

plt.show()
```

```python
# loss of the test set at each epoch
plt.plot(result['epoch'], result['test_cost'])
# plt.plot(result['epoch'][1:], result['test_cost'][1:])
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("loss of the train set")

plt.show()
```

```python
train_loss = np.array(result['train_cost'])
test_loss = np.array(result['test_cost'])
print("Train:", train_loss[:10])
print("Test:", test_loss[:10])
```
