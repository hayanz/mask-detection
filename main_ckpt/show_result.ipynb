```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
result = pd.read_csv('result.csv')
result.head()
```

```python
# show cost values of each set
fig = plt.figure()
ax1 = plt.subplot(1, 2, 1)
ax2 = plt.subplot(1, 2, 2)

ax1.plot(result['epoch'], result['train_cost'])  # train set
ax2.plot(result['epoch'], result['test_cost'])  # test set

plt.show()
```

```python
# show the accuracy of each set
fig = plt.figure()
ax1 = plt.subplot(1, 2, 1)
ax2 = plt.subplot(1, 2, 2)

ax1.plot(result['epoch'], result['train_acc'])  # train set
ax2.plot(result['epoch'], result['test_acc'])  # test set

plt.show()
```

```python
trains, tests = result['train_cost'], result['test_cost']
train_cost_min, test_cost_min = np.min(trains), np.min(tests)  # calculate

# print values with index
print("Train set:", train_cost_min, list(trains).index(train_cost_min))
print("Test set:", test_cost_min, list(tests).index(test_cost_min))
```
